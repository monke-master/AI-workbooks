sqrt(1 + exp(sqrt(x)) + cos(x)^2)/abs(1 - sin(x)^3) + ln(abs(2x))
–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import numpy as np
from matplotlib import pyplot as plt


def f(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x)**2)/np.abs(1 - np.sin(x)**3) + np.log(abs(2*x))

x = np.arange(20, 31, 1)
y = f(x)

half = x[:5]
half_y = f(half)


plt.grid()
plt.plot(x, y)
plt.show()

plt.scatter(half, half_y)


```
